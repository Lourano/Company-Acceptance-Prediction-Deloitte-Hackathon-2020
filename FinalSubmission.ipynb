```python
IsolateDict = {ord(c): f' {c} ' for c in SymbolsToIsolate}

RemoveDict = {ord(c): f'' for c in SymbolsToDelete}

def HandlePunctuation(Word):
    
    Word = Word.translate(RemoveDict)
    
    Word = Word.translate(IsolateDict)
    
    return Word
```

```python
ForTrain.AllFeatures = ForTrain.AllFeatures.apply(lambda x: HandlePunctuation(x))

ForTest.AllFeatures = ForTest.AllFeatures.apply(lambda x: HandlePunctuation(x))
```

```python
ForTrain.Target.value_counts(normalize = True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

WordVectorizer = TfidfVectorizer(analyzer = 'word', stop_words = 'english', ngram_range = (1, 2), lowercase = True,

                                 min_df = 5, max_features = 100000)
```

```python
from sklearn.model_selection import StratifiedKFold

KFold = StratifiedKFold(n_splits = 5, random_state = 666, shuffle = True)

TrainIndices = []

TestIndices = []

for TrainIndex, TestIndex in KFold.split(TrainSet, TrainSet.target):
    
    TrainIndices.append(TrainIndex)
    
    TestIndices.append(TestIndex)

TrainIndices = np.array(TrainIndices)

TestIndices = np.array(TestIndices)
```

```python
def WeightedAccuracy(YTrue, YPredicted):
    
    Weights = np.ones(YTrue.shape[0])
    
    Weights[YTrue == 2] = 2
    
    return Weights @ (YTrue == YPredicted).astype("int") / Weights.sum()
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

def TrainModelFirstPart(Algorithm):

    print('Start validation: ')
    
    if(Algorithm == 'LogisticRegression'):
    
        Classifier = LogisticRegression(penalty = 'l2', dual = False, tol = 0.0001, C = 1.3, fit_intercept = True, intercept_scaling = 1.0, 
        
        class_weight = None, random_state = 666)
        
    elif(Algorithm == 'ExtraTreesClassifier'):
        
        Classifier = ExtraTreesClassifier(n_estimators = 1000, max_depth = 10, n_jobs = -1, criterion = 'gini')
    
    Model = Pipeline([('WordVectorizer', WordVectorizer), ('Classifier', Classifier)])
    
    Metrics = []
        
    Fold = 0

    for TrainIdx, TestIdx in zip(TrainIndices, TestIndices):
        
        Fold += 1
        
        print("Fold:", Fold)
        
        Model.fit(ForTrain['AllFeatures'].iloc[TrainIdx], ForTrain['Target'].iloc[TrainIdx])
        
        print(Fold, 'Fitted')
        
        YPredicted = Model.predict(ForTrain['AllFeatures'].iloc[TestIdx])
    
        YTrue = ForTrain['Target'].iloc[TestIdx]
        
        Metrics.append(WeightedAccuracy(YTrue, YPredicted))
        
        print("Accuracy over fold:", Metrics[-1])
    
    print('Avarage Accuracy:', np.mean(Metrics))
    
    return YPredicted
```

```python
LRPredictions = TrainModelFirstPart('LogisticRegression')

print('LR fitted')

ETPredictions = TrainModelFirstPart('ExtraTreesClassifier')
```

```python
for TrainIdx, TestIdx in zip(TrainIndices, TestIndices):
            
    YTrue = ForTrain['Target'].iloc[TestIdx]

StackSet = pd.DataFrame({'LR' : LRPredictions, 'ET': ETPredictions, 'RealLabels': YTrue})

StackSet
```

```python
RealLabels = StackSet.RealLabels

StackSet = StackSet.drop(['RealLabels'], axis = 1)
```

```python
from xgboost import XGBClassifier

StackModel = XGBClassifier(n_estimators = 1000, max_depth = 20)

StackModel.fit(StackSet, RealLabels)

Results = []

Results.append(StackModel.predict_proba(StackSet))

Results = np.array(Results)

StackPredictions = pd.Series(Results.sum(axis = 0).argmax(-1))

RealLabels = np.array(RealLabels)

StackPredictions = np.array(StackPredictions)

print(WeightedAccuracy(RealLabels, StackPredictions))
```

```python
def TrainModelSecondPart(Algorithm):
    
    if(Algorithm == 'LogisticRegression'):
    
        Classifier = LogisticRegression(penalty = 'l2', dual = False, tol = 0.0001, C = 1.3, fit_intercept = True, intercept_scaling = 1.0, 
        
        class_weight = None, random_state = 666)
        
    elif(Algorithm == 'ExtraTreesClassifier'):
        
        Classifier = ExtraTreesClassifier(n_estimators = 1000, max_depth = 10, n_jobs = -1, criterion = 'gini')
    
    Model = Pipeline([('WordVectorizer', WordVectorizer), ('Classifier', Classifier)])
        
    Results = []
    
    Fold = 0

    for TrainIdx, TestIdx in zip(TrainIndices, TestIndices):
        
        Fold += 1
        
        print("Fold:", Fold)
        
        Model.fit(ForTrain['AllFeatures'].iloc[TrainIdx], ForTrain['Target'].iloc[TrainIdx])
        
        print(Fold, 'Fitted')
        
        Results.append(Model.predict_proba(ForTest['AllFeatures']))
        
    Results = np.array(Results)

    Prediction = pd.Series(Results.sum(axis = 0).argmax(-1))  
    
    return Prediction
```

```python
TestLRPredictions = TrainModelSecondPart('LogisticRegression')

print('LR fitted')

TestETPredictions = TrainModelSecondPart('ExtraTreesClassifier')

TestStackSet = pd.DataFrame({'LR': TestLRPredictions, 'ET': TestETPredictions})

Results = []

Results.append(StackModel.predict_proba(TestStackSet))

Results = np.array(Results)

Predictions = pd.Series(Results.sum(axis = 0).argmax(-1))

Predictions.value_counts()
```
